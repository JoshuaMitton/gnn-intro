```python
import random
random.seed(0)
import numpy as np
np.random.seed(0)
import os
import scipy
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import matplotlib.patches as mpatches
import matplotlib
from tqdm import tqdm

import scipy.sparse as sp

font = {'family' : 'sans-serif',
        'weight' : 'normal',
        'size'   : 22}

matplotlib.rc('font', **font)

if not os.path.exists('imgs'):
    os.makedirs('imgs')
```

## Karate club graph


```python
G = nx.karate_club_graph()
A = nx.adjacency_matrix(G)

club_bin = {}
club_color = []
for idx in nx.get_node_attributes(G, 'club'):
    if nx.get_node_attributes(G, 'club')[idx] == 'Mr. Hi':
        club_bin[idx] = 1
        club_color.append('r')
    else:
        club_bin[idx] = 0
        club_color.append('b')
nx.set_node_attributes(G,club_bin,'club_bin')

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()

pos = nx.spring_layout(G, seed=0)
nx.draw(G, pos=pos, ax=ax, width=2, node_size=500, node_color=club_color)

legend_elements = [Line2D([0], [0], marker='o', color='r', label='Mr. Hi', markersize=20, linewidth=0),
                   Line2D([0], [0], marker='o', color='b', label='Officer', markersize=20, linewidth=0)]
fig.legend(handles=legend_elements, prop={'size': 20}, loc='upper left')
fig.tight_layout()
plt.savefig('imgs/karate-club-graph.pdf')
plt.savefig('imgs/karate-club-graph.png')
plt.show()
```

## Gradient values 
$\nabla: \mathcal{F} (\mathcal{V}) \rightarrow \mathcal{F} (\mathcal{\varepsilon})$

$(\nabla f)_{ij} = f_i - f_j$

```python
adjacency = {}
club_bin = {}
club_color = []
for idx in nx.get_node_attributes(G, 'club'):
    if nx.get_node_attributes(G, 'club')[idx] == 'Mr. Hi':
        club_bin[idx] = 1
        club_color.append('r')
    else:
        club_bin[idx] = 0
        club_color.append('b')
nx.set_node_attributes(G,club_bin,'club_bin')

for i, row in enumerate(A):
    node_num = np.nonzero(row.T)[0].tolist()
    adj_nodes = np.array([G.nodes[node]['club_bin'] for node in node_num])
    for j, node in enumerate(adj_nodes):
        adjacency[(i, node_num[j])] = G.nodes[i]['club_bin'] - node

nx.set_edge_attributes(G,adjacency,name='grad')

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()

# pos = nx.spring_layout(G, k=2, seed=0)
pos = nx.spring_layout(G, seed=0)
labels = nx.get_edge_attributes(G,'grad')
# print(labels.values())

edge_colors = []
for key, val in labels.items():
    if val == -1:
        edge_colors.append('g')
    if val == 0:
        edge_colors.append('k')
    if val == 1:
        edge_colors.append('g')
# print(len(edge_colors))

nx.draw(G, pos, ax=ax, width=3, node_size=500, edge_color=edge_colors, node_color=club_color)


legend_elements = [Line2D([0], [0], marker='o', color='r', label='Mr. Hi', markersize=20, linewidth=0),
                   Line2D([0], [0], marker='o', color='b', label='Officer', markersize=20, linewidth=0)]
fig.legend(handles=legend_elements, prop={'size': 20}, loc='upper left')
fig.tight_layout()
plt.savefig('imgs/karate-club-gradient.pdf')
plt.savefig('imgs/karate-club-gradient.png')
plt.show()
```

$\nabla: \mathcal{F} (\mathcal{V}) \rightarrow \mathcal{F} (\mathcal{\varepsilon})$

$(\nabla f)_{e} = K^T f$

Where $K$ is an $(X,E)$ matrix, the incidence matrix, with X and E the number of nodes and edges respectively.

To construct $K$, for every edge, $e=(u,v)$, $K_{u,e} = +1$ and $K_{v,e} = -1$

```python
Adict = dict(A.todok().items())
K = np.zeros((A.shape[0], len(Adict)))

for i, edge in enumerate(Adict.keys()):
    K[edge[0],i] = 1
    K[edge[1],i] = -1
    
club_bin = {}
club_color = []
for idx in nx.get_node_attributes(G, 'club'):
    if nx.get_node_attributes(G, 'club')[idx] == 'Mr. Hi':
        club_bin[idx] = 1
        club_color.append('r')
    else:
        club_bin[idx] = 0
        club_color.append('b')
nx.set_node_attributes(G,club_bin,'club_bin')
f = np.array(list(club_bin.values()))

# print(K.T.shape)
# print(f.shape)

grad = K.T @ f
# print(grad.shape)
# print(grad)


fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()

pos = nx.spring_layout(G.to_directed(), seed=0)

edge_colors = []
for val in grad:
    if val == -1:
        edge_colors.append('g')
    if val == 0:
        edge_colors.append('k')
    if val == 1:
        edge_colors.append('g')
# print(len(edge_colors))
        
nx.draw(G.to_directed(), pos, ax=ax, width=3, node_size=500, edge_color=edge_colors, node_color=club_color)


legend_elements = [Line2D([0], [0], marker='o', color='r', label='Mr. Hi', markersize=20, linewidth=0),
                   Line2D([0], [0], marker='o', color='b', label='Officer', markersize=20, linewidth=0)]
fig.legend(handles=legend_elements, prop={'size': 20}, loc='upper left')
fig.tight_layout()
plt.savefig('imgs/karate-club-gradient-v2.pdf')
plt.savefig('imgs/karate-club-gradient-v2.png')
plt.show()
```

## Divegence values 
$div : \mathcal{F} (\mathcal{\varepsilon}) \rightarrow \mathcal{F} (\mathcal{V})$

$(\mathrm{div} F)_{i} = \dfrac{1}{a_i} \sum_{j:(i,j) \in \varepsilon} w_{ij} F_{ij}$

```python
nodes = {}
# A2 = np.diagonal(np.multiply(A@A.todense(),np.identity(A.shape[0])))
# A2 = A2/np.max(A2)
A2 = np.array(A.todense().sum(axis=1).reshape(-1))
# A2 = A2/np.max(A2)
A2 = A2[0]
# print(A2)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()

cmap=plt.cm.viridis
cmap=plt.cm.magma
nx.draw(G,pos,ax=ax, width=2, node_size=500, node_color=A2,cmap=cmap)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin = np.min(A2), vmax=np.max(A2)))
sm._A = []
plt.colorbar(sm)

fig.tight_layout()
plt.savefig('imgs/karate-club-divengence.pdf')
plt.savefig('imgs/karate-club-divengence.png')
plt.show()

A2 = A2/np.max(A2)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()

cmap=plt.cm.viridis
cmap=plt.cm.magma
nx.draw(G,pos,ax=ax, width=2, node_size=500, node_color=A2,cmap=cmap)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin = 0, vmax=1))
sm._A = []
plt.colorbar(sm)

fig.tight_layout()
plt.savefig('imgs/karate-club-divengence-norm.pdf')
plt.savefig('imgs/karate-club-divengence-norm.png')
plt.show()
```

$div : \mathcal{F} (\mathcal{\varepsilon}) \rightarrow \mathcal{F} (\mathcal{V})$

$(div)_{i} = K f$

Where $K$ is an $(X,E)$ matrix, the incidence matrix, with X and E the number of nodes and edges respectively and f is the fucntion over edges, constant in this instance.

To construct $K$, for every edge, $e=(u,v)$, $K_{u,e} = +1$ and $K_{v,e} = -1$

```python
# print(nx.adjacency_matrix(G.to_directed()))

Adict = dict(A.todok().items())
K = np.zeros((A.shape[0], len(Adict)))

f = []
for key, item in Adict.items():
    if key[0]>key[1]:
        f.append(1)
    else:
        f.append(1)
f = np.array(f)
# print(f)

for i, edge in enumerate(Adict.keys()):
    K[edge[0],i] = 1
    K[edge[1],i] = 1
# print(K)
    
# f = np.ones(K.shape[1])

# print(K.shape)
# print(f.shape)

div = np.abs(K @ f)
# div = div/2
# print(div.shape)
# print(div)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()

cmap=plt.cm.viridis
cmap=plt.cm.magma
nx.draw(G,pos,ax=ax, width=2, node_size=500, node_color=div,cmap=cmap)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin = np.min(div), vmax=np.max(div)))
sm._A = []
plt.colorbar(sm)

fig.tight_layout()
plt.savefig('imgs/karate-club-divengence-v2.pdf')
plt.savefig('imgs/karate-club-divengence-v2.png')
plt.show()

div = div/np.max(div)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()

cmap=plt.cm.viridis
cmap=plt.cm.magma
nx.draw(G,pos,ax=ax, width=2, node_size=500, node_color=div,cmap=cmap)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin = 0, vmax=1))
sm._A = []
plt.colorbar(sm)

fig.tight_layout()
plt.savefig('imgs/karate-club-divengence-norm-v2.pdf')
plt.savefig('imgs/karate-club-divengence-norm-v2.png')
plt.show()
```

## Laplacian values 
$\Delta : \mathcal{F} (\mathcal{V}) \rightarrow \mathcal{F} (\mathcal{V})$

$(\Delta F)_{i} = \dfrac{1}{a_i} \sum_{j:(i,j) \in \varepsilon} w_{ij} (f_i - f_j)$

```python
lapl = np.zeros(A.shape[0])
club_bin = {}
club_color = []
for idx in nx.get_node_attributes(G, 'club'):
    if nx.get_node_attributes(G, 'club')[idx] == 'Mr. Hi':
        club_bin[idx] = 1
        club_color.append('r')
    else:
        club_bin[idx] = 0
        club_color.append('b')
nx.set_node_attributes(G,club_bin,'club_bin')

for i, row in enumerate(A):
    node_num = np.nonzero(row.T)[0].tolist()
    adj_nodes = np.array([G.nodes[node]['club_bin'] for node in node_num])
    for j, node in enumerate(adj_nodes):
        lapl[i] += G.nodes[i]['club_bin'] - node


fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()
pos = nx.spring_layout(G, seed=0)
cmap=plt.cm.magma
nx.draw(G, pos, ax=ax, width=2, node_size=500, node_color=lapl, cmap=cmap)
# nx.draw_networkx_labels(G,pos,labels=labels,font_size=12)
vmin = min(lapl)
vmax = max(lapl)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
plt.colorbar(sm)
plt.savefig('imgs/karate-club-laplacian-sumovergrads.pdf')
plt.savefig('imgs/karate-club-laplacian-sumovergrads.png')
plt.show()

lapl = lapl/np.max(lapl)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()
pos = nx.spring_layout(G, seed=0)
cmap=plt.cm.magma
# matplotlib.colors.Normalize(vmin=0, vmax=1)
vmin = -1 #min(lapl)
vmax = 1 #max(lapl)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
nx.draw(G, pos, ax=ax, width=2, node_size=500, node_color=lapl, cmap=cmap, vmin=vmin, vmax=vmax)
# nx.draw_networkx_labels(G,pos,labels=labels,font_size=12)

sm._A = []
plt.colorbar(sm)
plt.savefig('imgs/karate-club-laplacian-sumovergrads-norm.pdf')
plt.savefig('imgs/karate-club-laplacian-sumovergrads-norm.png')
plt.show()
```

## Laplacian values 

$\Delta : \mathcal{F} (\mathcal{V}) \rightarrow \mathcal{F} (\mathcal{V})$

$(\Delta F)_{i} = div(grad(f)) = K K^{T} f$

Where $K$ is an $(X,E)$ matrix, the incidence matrix, with X and E the number of nodes and edges respectively.

To construct $K$, for every edge, $e=(u,v)$, $K_{u,e} = +1$ and $K_{v,e} = -1$

Lets look at $K K^{T}$

```python
Adict = dict(A.todok().items())
K = np.zeros((A.shape[0], len(Adict)))

for i, edge in enumerate(Adict.keys()):
    K[edge[0],i] = 1
    K[edge[1],i] = -1
    
L = K@K.T
# print(L)

club_bin = {}
club_color = []
for idx in nx.get_node_attributes(G, 'club'):
    if nx.get_node_attributes(G, 'club')[idx] == 'Mr. Hi':
        club_bin[idx] = 1
        club_color.append('r')
    else:
        club_bin[idx] = 0
        club_color.append('b')
nx.set_node_attributes(G,club_bin,'club_bin')
f = np.array(list(club_bin.values()))
# print(f)

lapl = L@f
# print(lapl)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()
pos = nx.spring_layout(G, seed=0)
cmap=plt.cm.magma
vmin = min(lapl)
vmax = max(lapl)
nx.draw(G, pos, ax=ax, width=2, node_size=500, node_color=lapl, cmap=cmap, vmin=vmin, vmax=vmax)
# nx.draw_networkx_labels(G,pos,labels=labels,font_size=12)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
plt.colorbar(sm)
plt.savefig('imgs/karate-club-laplacian-incidence.pdf')
plt.savefig('imgs/karate-club-laplacian-incidence.png')
plt.show()

lapl = lapl/np.max(lapl)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()
pos = nx.spring_layout(G, seed=0)
cmap=plt.cm.magma
# matplotlib.colors.Normalize(vmin=0, vmax=1)
vmin = -1 #min(lapl)
vmax = 1 #max(lapl)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
nx.draw(G, pos, ax=ax, width=2, node_size=500, node_color=lapl, cmap=cmap, vmin=vmin, vmax=vmax)
# nx.draw_networkx_labels(G,pos,labels=labels,font_size=12)

sm._A = []
plt.colorbar(sm)
plt.savefig('imgs/karate-club-laplacian-incidence-norm.pdf')
plt.savefig('imgs/karate-club-laplacian-incidence-norm.png')
plt.show()

```

## Laplacian values 
<!-- $\Delta = \mathrm{A^{-1}} (\mathrm{D}-\mathrm{W})$ -->

In spectral graph theory the graph Laplacian is refered to as follows. This definition is the Laplacian operator which when applied to the function over graph nodes yields the same result as the Laplacian definition above.

Laplacian : $\Delta = (\mathrm{D}-\mathrm{A})$

```python
club_bin = {}
club_color = []
for idx in nx.get_node_attributes(G, 'club'):
    if nx.get_node_attributes(G, 'club')[idx] == 'Mr. Hi':
        club_bin[idx] = 1
        club_color.append('r')
    else:
        club_bin[idx] = 0
        club_color.append('b')
nx.set_node_attributes(G,club_bin,'club_bin')
f = np.array(list(club_bin.values()))
# print(f)

D = np.multiply(A.sum(axis=1).reshape(-1), np.identity(A.shape[0]))
lapl = D-A.todense()
lapl_op = lapl
# print(lapl)
np.savetxt('imgs/karate-club-laplacian-matrix-spectral.txt', lapl)

lapl = (lapl@f).tolist()
lapl = np.array([float(i) for i in lapl[0]])

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()
pos = nx.spring_layout(G, seed=0)
cmap=plt.cm.magma
# nx.draw_networkx_labels(G,pos,labels=labels,font_size=12)
vmin = min(lapl)
vmax = max(lapl)
nx.draw(G, pos, ax=ax, width=2, node_size=500, node_color=lapl, cmap=cmap, vmin=vmin, vmax=vmax)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
plt.colorbar(sm)
plt.savefig('imgs/karate-club-laplacian-spectral.pdf')
plt.savefig('imgs/karate-club-laplacian-spectral.png')
plt.show()

lapl = lapl/np.max(lapl)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()
pos = nx.spring_layout(G, seed=0)
cmap=plt.cm.magma
# nx.draw_networkx_labels(G,pos,labels=labels,font_size=12)
vmin = -1
vmax = 1
nx.draw(G, pos, ax=ax, width=2, node_size=500, node_color=lapl, cmap=cmap, vmin=vmin, vmax=vmax)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
plt.colorbar(sm)
plt.savefig('imgs/karate-club-laplacian-spectral-norm.pdf')
plt.savefig('imgs/karate-club-laplacian-spectral-norm.png')
plt.show()
```

From the three different ways of defining the Laplacian on a graph it can be seen that they evaluate to the same result. This provides a nice bridge between the more intuitive definition given by $(\Delta F)_{i} = \dfrac{1}{a_i} \sum_{j:(i,j) \in \varepsilon} w_{ij} (f_i - f_j)$ which is easy to see the derivation from the traditional Laplacian into a graph domain and the definition given by spectral graph theory, $\Delta = (\mathrm{D}-\mathrm{A})$, which is widely used in graph neural networks. Finally, the operators generated using the spectral graph theory and incidence matrix methods can be compared (and hopefully they yield the same operator).

```python
## Spectral Laplcaian operator
D = np.multiply(A.sum(axis=1).reshape(-1), np.identity(A.shape[0]))
lapl = D-A.todense()

## incidence matrix Laplacian operator
Adict = dict(A.todok().items())
K = np.zeros((A.shape[0], len(Adict)))
for i, edge in enumerate(Adict.keys()):
    K[edge[0],i] = 1
    K[edge[1],i] = -1  

L = K@K.T

## Difference between the two operators
print(f'Difference between the spectral and incidence operators : {np.sum(lapl-L)}')
```
